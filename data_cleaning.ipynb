# Airbnb Europe Pricing

Objective: Explore Airbnb listing data across 10 European cities to understand pricing patterns and build a price prediction model later.

Dataset: 20 CSV files, 10 cities, weekday/weekend file per city

This notebook:
Opens each CSV file and adds a city/weekday-weekend label
Combines everything into one table
Checks columns, data types, and missing values
Removes duplicates, fixes data types, and removes price outliers
Saves the dataset as airbnb_cleaned.csv

```python
import pandas as pd
import numpy as np

#import datasets
amsterdam_weekdays = pd.read_csv('../airbnb/data/amsterdam_weekdays.csv', index_col=0)
amsterdam_weekends = pd.read_csv('../airbnb/data/amsterdam_weekends.csv', index_col=0)

athens_weekdays = pd.read_csv('../airbnb/data/athens_weekdays.csv', index_col=0)
athens_weekends = pd.read_csv('../airbnb/data/athens_weekends.csv', index_col=0)

barcelona_weekdays = pd.read_csv('../airbnb/data/barcelona_weekdays.csv', index_col=0)
barcelona_weekends = pd.read_csv('../airbnb/data/barcelona_weekends.csv', index_col=0)

berlin_weekdays = pd.read_csv('../airbnb/data/berlin_weekdays.csv', index_col=0)
berlin_weekends = pd.read_csv('../airbnb/data/berlin_weekends.csv', index_col=0)

budapest_weekdays = pd.read_csv('../airbnb/data/budapest_weekdays.csv', index_col=0)
budapest_weekends = pd.read_csv('../airbnb/data/budapest_weekends.csv', index_col=0)

lisbon_weekdays = pd.read_csv('../airbnb/data/lisbon_weekdays.csv', index_col=0)
lisbon_weekends = pd.read_csv('../airbnb/data/lisbon_weekends.csv', index_col=0)

london_weekdays = pd.read_csv('../airbnb/data/london_weekdays.csv', index_col=0)
london_weekends = pd.read_csv('../airbnb/data/london_weekends.csv', index_col=0)

paris_weekdays = pd.read_csv('../airbnb/data/paris_weekdays.csv', index_col=0)
paris_weekends = pd.read_csv('../airbnb/data/paris_weekends.csv', index_col=0)

rome_weekdays = pd.read_csv('../airbnb/data/rome_weekdays.csv', index_col=0)
rome_weekends = pd.read_csv('../airbnb/data/rome_weekends.csv', index_col=0)

vienna_weekdays = pd.read_csv('../airbnb/data/vienna_weekdays.csv', index_col=0)
vienna_weekends = pd.read_csv('../airbnb/data/vienna_weekends.csv', index_col=0)
```

### Add city and weekday/weekend labels to each dataframe

```python
amsterdam_weekdays['city'] = 'amsterdam'
amsterdam_weekdays['price_period'] = 'weekday'
amsterdam_weekends['city'] = 'amsterdam'
amsterdam_weekends['price_period'] = 'weekend'

athens_weekdays['city'] = 'athens'
athens_weekdays['price_period'] = 'weekday'
athens_weekends['city'] = 'athens'
athens_weekends['price_period'] = 'weekend'

barcelona_weekdays['city'] = 'barcelona'
barcelona_weekdays['price_period'] = 'weekday'
barcelona_weekends['city'] = 'barcelona'
barcelona_weekends['price_period'] = 'weekend'

berlin_weekdays['city'] = 'berlin'
berlin_weekdays['price_period'] = 'weekday'
berlin_weekends['city'] = 'berlin'
berlin_weekends['price_period'] = 'weekend'

budapest_weekdays['city'] = 'budapest'
budapest_weekdays['price_period'] = 'weekday'
budapest_weekends['city'] = 'budapest'
budapest_weekends['price_period'] = 'weekend'

lisbon_weekdays['city'] = 'lisbon'
lisbon_weekdays['price_period'] = 'weekday'
lisbon_weekends['city'] = 'lisbon'
lisbon_weekends['price_period'] = 'weekend'

london_weekdays['city'] = 'london'
london_weekdays['price_period'] = 'weekday'
london_weekends['city'] = 'london'
london_weekends['price_period'] = 'weekend'

paris_weekdays['city'] = 'paris'
paris_weekdays['price_period'] = 'weekday'
paris_weekends['city'] = 'paris'
paris_weekends['price_period'] = 'weekend'

rome_weekdays['city'] = 'rome'
rome_weekdays['price_period'] = 'weekday'
rome_weekends['city'] = 'rome'
rome_weekends['price_period'] = 'weekend'

vienna_weekdays['city'] = 'vienna'
vienna_weekdays['price_period'] = 'weekday'
vienna_weekends['city'] = 'vienna'
vienna_weekends['price_period'] = 'weekend'
```

```python
#merge
df = pd.concat([
    amsterdam_weekdays, amsterdam_weekends,
    athens_weekdays, athens_weekends,
    barcelona_weekdays, barcelona_weekends,
    berlin_weekdays, berlin_weekends,
    budapest_weekdays, budapest_weekends,
    lisbon_weekdays, lisbon_weekends,
    london_weekdays, london_weekends,
    paris_weekdays, paris_weekends,
    rome_weekdays, rome_weekends,
    vienna_weekdays, vienna_weekends
], ignore_index=True)

df.shape
```

### Data cleaning

```python
#missing value
df.isnull().any()
```

```python
# remove duplicate
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape:", df.shape)
```

```python
# bedrooms should be whole number
df['bedrooms'] = df['bedrooms'].astype(int) 
df['city'] = df['city'].astype('category')
df['room_type'] = df['room_type'].astype('category')
df['price_period'] = df['price_period'].astype('category')

df.dtypes
```

```python
df['realSum'].describe()
```

```python
# max is way higher than the 75th percentile, 
# price column has some expensive listings that might distort the charts and the model later
# use the IQR rule to cut them off.
Q1 = df['realSum'].quantile(0.25)
Q3 = df['realSum'].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR
print("Upper limit:", upper_limit)

before = len(df)
df = df[df['realSum'] <= upper_limit]
after = len(df)
print(f"Removed {before - after} outlier rows")
print("New shape:", df.shape)
```

```python
# save
df.to_csv('../airbnb/data/clean_data/airbnb_cleaned.csv', index=False)
print("Saved:", df.shape)
```
